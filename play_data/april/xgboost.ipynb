```python
label_col = df['Label']
```

```python
df.drop(['Weight','KaggleSet','Label'], axis=1, inplace=True)
```

```python
df.rename(columns={'KaggleWeight' : 'Weight'}, inplace=True)
```

```python
# s -> signal(신호) , b -> background(잡음)
label_col.replace(('s', 'b'), (1,0), inplace=True)
```

```python
X = df.iloc[ : , 1 : ] 
y = label_col
```

```python
import xgboost as xgb
```

```python
X
```

```python
y.values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,pd.DataFrame(y.values.reshape(-1,1)),test_size=0.2 , random_state=42)
```

```python
y_test.value_counts()
```

```python
y_train
```

```python
xgbmat = xgb.DMatrix(X_train, y_train, missing=-999.0)
```

```python
param = {
    'objective' : 'binary:logitraw',
    'eta' : 0.1 ,
    'max_depth' : 6,
    'eval_metric' : 'auc',
}
```

```python
bst = xgb.train(param,xgbmat,120,[(xgbmat, 'train')])
```

```python
xgbmat_test = xgb.DMatrix(X_train, y_train, missing=-999.0)
```

```python
y_hat = bst.predict(xgbmat_test)
```

```python
from lightgbm import LGBMClassifier
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

dataset = load_breast_cancer()


ftr = dataset.data
target = dataset.target

X_train, X_test, y_train, y_test = train_test_split(ftr, target, test_size=0.2 , random_state=156)


gbm = LGBMClassifier(n_estimators=400)

gbm.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='logloss',
        eval_set=[(X_test, y_test)], verbose=True)
```
