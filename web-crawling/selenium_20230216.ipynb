```python
from bs4 import BeautifulSoup as BS
from selenium.webdriver.common.by import By
def get_lyrics(driver,tag,class_id,selector_id):
    bs = BS(driver.page_source) # HTML return
    bs.find(tag, {'class' : f"{class_id}"})
    driver.find_element(By.CSS_SELECTOR, selector_id).click()
    bs = BS(driver.page_source)
    lyrics = bs.find("div", {'class': "lyrics"})
    return lyrics.find("p").text
```

```python
"""
kbo --> bmi value
"""
import re
from collections import defaultdict
def get_bmi(kbo):
    bmi_dict = defaultdict(list)
    for person in kbo:
        p = re.compile("[0-9]+")
        h,w = p.findall(person['신장/체중'])
        # h,w = person['신장/체중'].split('/')
        h = int(h.replace("cm",""))
        w = int(w.replace("kg",""))
        age = 2024-int(person['생년월일'][:4])
        bmi_dict[person['선수명']].append((round(w/(h/100)**2,1),age))
    return bmi_dict

```

```python
bmi_dict = get_bmi(kbo)
bmi_dict
```

```python
bmi_top5 = sorted(get_bmi(kbo).items(),key= lambda x:max(x[1])[0],reverse=True)[:5]
for name,x in bmi_top5:
    print(f'name:{name}, bmi:{x[0][0]}, age:{x[0][1]}')
```

```python
import pickle
from datetime import date


with open("./kbo_player.pkl", "rb") as f:
    kbo = pickle.load(f)
for play in kbo:
    height, weight = play['신장/체중'].split("/")
    bmi = int(weight[:-2]) / (int(height[:-2]) / 100) ** 2
    play['BMI'] = bmi
    play['나이'] = date.today().year - int(play['생년월일'][:4])


sorted({"-".join([x['선수명'],x['등번호']]) : x['BMI'] for x in kbo}.items(), key=lambda x: x[1], reverse=True)

salary = {}
for play in kbo:
    if play['연봉'][-2:] == '만원':
        if play['Team'] in salary:
            salary[play['Team']] += [int(play['연봉'][:-2])]
        else:
            salary[play['Team']] = [int(play['연봉'][:-2])]

sorted({key : ( sum(value)/len(value)) for key,value in salary.items() }.items(),
       key=lambda x : x[1], reverse=True)
```

```python
import numpy as np
sorted({key : (np.median(value)) for key,value in salary.items() }.items(),
       key=lambda x : x[1], reverse=True)
```

# 고속도로 휴게소 데이터

#### xml --> json 변환

```python
import requests,xml.etree.ElementTree as ET

def xml_to_json():
    total = []

    for i in range(1,4):
        url = f"http://data.ex.co.kr/openapi/locationinfo/locationinfoRest?key=8097256368&type=xml&numOfRows=1000&pageNo={str(i)}"
        r = requests.get(url)
        root = ET.fromstring(r.text)
        items = root.iter(tag='list')

        for element in items:
            tmp = {}
            for x in element:
                tmp[x.tag] = x.text
            total.append(tmp)
    return total
```

```python
total = xml_to_json()
```
